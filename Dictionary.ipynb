```python
books = {
    "BookName_1" : "The Alchemist", 
    "Price_1": 2199,
    "BookName_2" : "Rich Dad Poor Dad", 
    "Price_2": 4299,
    "BookName_3" : "Think and Grow Rich", 
    "Price_3": 3399,
    "BookName_4" : "The Power of Subconscious Mind",
    "Price_4": 2499
}
print("The details of Books before adding new book is :",books)


books["BookName_5"] = "Atomic Habits"
books["Price"] = 3199

print("The details of Books after adding a new book is :",books)

print("Deleting a book")
del(books["BookName_5"])
print("Updated book dictionary :",books)
del(books["Price"])
```

```python
## POP FUNCTION
removed_price = books.pop("Price_1")
print(books)
print(removed_price)
```

```python
fav_colors = {
    'john':'blue',
    'gabe':'orange',
    'mary':'yellow',
    'mike':'purple',
    'lucas':'yellow',
    'sarah':'green'
}
print(fav_colors)
fav_colors_list = fav_colors.items()
print(fav_colors_list)

print("Fetching the key value pairs using for loop:")
for key, value in fav_colors.items():
    print(f"{key} -> {value}")

print("Fetching only the keys")
fav_colors_keys = fav_colors.keys()
print(fav_colors_keys)   

print("Fetching only values")
fav_colors_values = fav_colors.values()
print(fav_colors_values)
```

```python
for value in fav_colors.values():
    print(value)
```

```python
## Remvoving duplicate values from the values
print("Therse are the colors present in the dictionary")
for value in fav_colors.values():
    print(value)

print("Removing duplicate color values from the dictionary")
unique_colors = set(fav_colors.values())
print(unique_colors)
```
