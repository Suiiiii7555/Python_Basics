#### 1. Write a program to dynamically add a function to a class methods

```python
import types

class Student:
    def __init__(self, name, age, rollno):
        self.name = name
        self.age = age
        self.rollno = rollno

    def display(self):
        return f"Name: {self.name}  Age: {self.age}  Roll number: {self.rollno}"

def setAge(self, age):
    self.age = age


student1 = Student("Sathvik", 23, 1)

student1.setAge = types.MethodType(setAge, student1)

student1.setAge(24)

print(student1.display())
```

#### 2. Write a program to dynamically delete a function from a class method

```python
import types

class Car:
    def porsche(self):
        return "Porsche 911 here"

    def BMW(self):
        return "BMW M5 CS here"


def maruti(self):
    return "Maruti Suzuki Swift here"

car1 = Car()

car1.maruti = types.MethodType(maruti, car1)
print(car1.maruti())

delattr(car1, 'maruti')

if hasattr(car1, 'maruti'):
    print(car1.maruti())
else:
    print("maruti method has been deleted dynamically")
```

#### 3. Write a program to delete a class method (inside class)

```python
import types

class Car:
    def porsche(self):
        return "Porsche 911 here"

    def BMW(self):
        return "BMW M5 CS here"
    
    def Ferrari(self):
        return "Ferrari F8 Tributo here"


def maruti(self):
    return "Maruti Suzuki Swift here"

car1 = Car()

car1.maruti = types.MethodType(maruti, car1)
print(car1.maruti())

print(car1.Ferrari())
del car1.Ferrari
if hasattr(car1, 'Ferrari'):
    print(car1.Ferrari())
else:
    print("Ferrari method has been deleted dynamically")
```
