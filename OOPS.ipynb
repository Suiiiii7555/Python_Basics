#### 1. Write a program to dynamically add a function to a class methods

```python
import types

class Student:
    def __init__(self, name, age, rollno):
        self.name = name
        self.age = age
        self.rollno = rollno

    def display(self):
        return f"Name: {self.name}  Age: {self.age}  Roll number: {self.rollno}"

def setAge(self, age):
    self.age = age


student1 = Student("Sathvik", 23, 1)

student1.setAge = types.MethodType(setAge, student1)

student1.setAge(24)

print(student1.display())
```

#### 2. Write a program to dynamically delete a function from a class method

```python
import types

class Car:
    def porsche(self):
        return "Porsche 911 here"

    def BMW(self):
        return "BMW M5 CS here"


def maruti(self):
    return "Maruti Suzuki Swift here"

car1 = Car()

car1.maruti = types.MethodType(maruti, car1)
print(car1.maruti())

delattr(car1, 'maruti')

if hasattr(car1, 'maruti'):
    print(car1.maruti())
else:
    print("maruti method has been deleted dynamically")
```

```python
class BankAccount:
    # Class variable to manage a universal discount rate for all accounts
    discount_rate = 0.05  # 5% discount across all accounts
 
    def __init__(self, owner, balance=0):
        self.owner = owner  # Instance variable for account holder's name
        self.balance = balance  # Instance variable for account balance
 
    def deposit(self, amount):
        """Instance method to deposit money into the account."""
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")
 
    def withdraw(self, amount):
        """Instance method to withdraw money from the account."""
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")
 
    @classmethod
    def set_discount_rate(cls, new_discount_rate):
        """Class method to update the bank-wide discount rate."""
        if 0 <= new_discount_rate <= 1:
            cls.discount_rate = new_discount_rate
            print(f"New bank-wide discount rate: {cls.discount_rate * 100}%")
        else:
            print("Invalid discount rate. Must be between 0 and 1.")
 
    @staticmethod
    def calculate_interest(balance, rate=0.03):
        """Static method to calculate interest on a given balance."""
        interest = balance * rate
        return interest
 
 
# Usage of the BankAccount class:
 
# Creating account instances
account1 = BankAccount("Alice", 1000)
account2 = BankAccount("Bob", 1500)
 
# Instance method usage (deposit and withdraw)
account1.deposit(500)  # Alice deposits 500
account1.withdraw(200)  # Alice withdraws 200
 
# Class method usage (set discount rate for all accounts)
BankAccount.set_discount_rate(0.1)  # Bank-wide discount rate set to 10%
 
# Static method usage (calculate interest for an account)
interest = BankAccount.calculate_interest(account2.balance, 0.04)
print(f"Interest on Bob's account: {interest}")  # Calculate interest for Bob
 
# Creating another account and checking how the discount applies
account3 = BankAccount("Charlie", 2000)
print(f"Charlie's account discount rate: {BankAccount.discount_rate * 100}%")
 
```

#### 4. Write a code for Mutilevel inheritance and check if super works over here

```python
# Base class
class Person:
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("Name:", self.name)


# Intermediate class
class Employee(Person):
    def __init__(self, name, emp_id):
        super().__init__(name)     # Call Person constructor
        self.emp_id = emp_id

    def show_employee_details(self):
        print("Employee ID:", self.emp_id)


# Derived class
class Manager(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)   # Call Employee constructor
        self.department = department

    def show_manager_details(self):
        print("Department:", self.department)


# Driver code
m1 = Manager("Rahul", 101, "IT")

m1.show_name()                 # From Person
m1.show_employee_details()     # From Employee
m1.show_manager_details()      # From Manager
```
