```python
coupons = {"RAM","PSU","GPU","Chassis"}
store = {"CPU","RAM","GPU","HDD","SSD"}

## To check which all coupons are with me 
print(coupons.intersection(store))

## To check which coupons are not with me
print("The coupons you are missing out are :",store.difference(coupons)) # or another way to represent -> print(store - coupons)
print("The combination of coupons that you have and the store to which you can't go :",coupons.symmetric_difference(store)) # or another way to represent -> print(coupons - store)
```
