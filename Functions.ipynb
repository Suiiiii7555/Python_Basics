```python
def log(val,decor):
    val = str(val)
    val_len = len(val)
    prefix = decor*4 + decor*val_len + decor*2
    print(prefix)
    print(decor * 2 + " " + val + " " + decor * 2)
    print(prefix)
log(102,'*')
log(9,'*')
log("Sathvik","-_-") ## This is called a positional argument type of function
log(decor="*",val="Aag") ## This is called a keyword argument type of function
```

```python
def argument(*values,decor="-"):
    for val in values:
        val = str(val)
        val_len = len(val)
        prefix = decor*2 + decor*val_len + decor*2
        print(prefix)
        print(decor + " " + str(val) + " " + decor)
        print(prefix)
argument(45,50,55,60)
```

```python
def argument(*values,decor="-", end="\n"):
    for val in values:
        val = str(val)
        val_len = len(val)
        prefix = decor*2 + decor*val_len + decor*2
        print(prefix)
        print(decor + " " + str(val) + " " + decor)
        print(prefix, end=end)
argument(45,50,55,60,end="\n\n")
```

```python
def argument_1(*values,decor="-", end="\n", padding=2):
    for val in values:
        val = str(val)
        val_len = len(val)
        prefix = (decor*padding + decor*val_len + decor*padding)
        output = (decor * (padding - 1) + " " + str(val) + " " + decor * (padding - 1))
        print(prefix)
        print(output)
        print(prefix, end=end)
argument_1(12,16,decor="~",padding=4)
```
