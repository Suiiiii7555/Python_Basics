## LIST

```python
grocery = ["Milk","Bread","Egg","Cucumber","Onion","Potato","Chillies","Corriander","Tomato"]
print(grocery[0])
print(grocery[::-1])
print("Before updating",grocery)
grocery[2] = "Banana"
print("After updating",grocery)
grocery.remove("Onion")
print("List after removal operation is",grocery)
removed_item = grocery.pop(2)
print(f"List after removal of [{removed_item}] is {grocery}")
grocery.append(removed_item)
print("List after adding back the removed item is",grocery)
length = len(grocery)
print("Total elements in the grocery list are :",length)
```

```python
numbers = [1,4,7,3,7,11,2,5]
ascending = sorted(numbers)
descending = sorted(numbers,reverse=True)
print("Sorted in ascending order :",ascending)
print("Sorted in descending order :",descending)
```

```python
for i in numbers:
    print(i)
numbers_squared = [i**2 for i in numbers]
print("Numbers squared are :",numbers_squared)
```

#### Slicing of Lists ####

```python
print(grocery)
print(grocery[::2])
print(dir(grocery))
print("Apple" in grocery)
```
