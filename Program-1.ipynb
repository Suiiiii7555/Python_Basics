#### Validate User input

```python
def check_length(user_name):
    if len(user_name) < 12:
        print("Username must be at least 12 characters long")
    else:
        print("User name can be accepted")

def check_spaces(user_name):
    if ' ' in user_name:
        print("Username must not contain spaces")
    else:
        check_length(user_name)

def check_digits(user_name):
    if user_name.isalpha():
        print("Username can be accepted")
    else:
        check_spaces(user_name)

user_name = input("Enter your name : ")
check_length(user_name)
check_spaces(user_name)
check_digits(user_name)
```

#### 2. Compound Interest Calculator

```python
principle = 0
time = 0
rate = 0

while principle <= 0:
    principle = float(input("Enter the principle amount : "))
    if principle <= 0:
        print("Principle cannot be less than or equal to 0")
print("Your principle amount is :",principle)

while rate <= 0:
    rate = float(input("Enter the interest rate : "))
    if rate <= 0:
        print("Interest rate cannot be less than or equal to 0")
print("The rate of interest is :",rate)

while time <= 0:
    time = float(input("Enter the time in terms of years : "))
    if time <= 0:
        print("Years cannot be less than or equal to 0")
print("The tenure is :",time)

total = principle * pow((1+rate/100),time)
print(f"The total amount after {time} years is : Rs {total:.2f}")
```

#### Count down program

```python
import time 
my_time = int(input("Enter the time in seconds : "))
for i in range(my_time,0,-1):
    seconds = i % 60
    minutes = int(i / 60) % 60
    hour = int(i / 3600)
    print(f"{hour:02}:{minutes:02}:{seconds:02}")
    time.sleep(1)
print("Time's UP!")
```
