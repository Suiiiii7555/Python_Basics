```python
## CREATING A STACK

from collections import deque
cookies = deque((1,2,3,4))
# cookies.appendleft("Ragi Cookies")

print(cookies)

first_cookie = cookies.popleft()
print("First cookie served:", first_cookie)
print("Cookies left in the stack:", cookies, len(cookies))
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n * factorial(n-1)
num = 5
result = factorial(num)
print("Factorial of", num, "is", result)
```
